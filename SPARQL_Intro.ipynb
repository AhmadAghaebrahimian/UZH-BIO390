**SPARQL**

1. Prefix declarations: Abbreviating URIs
          PREFIX foo: <http://example.com/resources/>

2. Result clause: Identify what information to return
          SELECT *

3. Dataset definition: Which RDF graph(s) to query
          FROM

4. Query pattern: Specify what to query
          WHERE {
                  ?s ?p ?o .
                }

5. Modifiers, Slicing, Ordering, or Rearranging the results
          ORDER BY ?s


Open Dbpedia Sparql endpoint https://dbpedia.org/sparql

      PREFIX rdf:	<http://www.w3.org/1999/02/22-rdf-syntax-ns#>
      
      SELECT DISTINCT ?concept
      WHERE {
          ?subject rdf:type ?concept .
      } 
      ORDER BY DESC(?concept)
      LIMIT 10
      OFFSET 2


      PREFIX dbp-owl: <http://dbpedia.org/ontology/>
      PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
      
      SELECT DISTINCT ?country ?label 
      WHERE {
        ?country a dbp-owl:Country .        
        ?country rdfs:label ?label .           
        FILTER langMatches(lang(?label),"en")  
        FILTER contains(?label,"land")         
      }


Built-in filter functions

    Logical: !, &&, ||
    Math: +, -, *, /
    Comparison: =, !=, >, <, ...
    SPARQL tests: isURI, isBlank, isLiteral, bound
    SPARQL accessors: str, lang, datatype
    Other: sameTerm, langMatches, regex

Open UniProt Sparql endpoint: https://sparql.uniprot.org/

    Shortest comments for a UniProtKB Natural Variant Annotation

      PREFIX up: <http://purl.uniprot.org/core/>
      PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>

      SELECT ?annotation ?comment
      WHERE {
        ?annotation a up:Natural_Variant_Annotation ;
                    rdfs:comment ?comment . 
      } 
      ORDER BY (STRLEN(?comment))

    
    
    Find how often an article in pubmed was used in an evidence tag

      PREFIX up: <http://purl.uniprot.org/core/>
      PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
      
      SELECT ?source (COUNT(?attribution) AS ?attribitions)
      WHERE
      {
        ?protein  a up:Protein ;
                  up:annotation ?annotation .
        ?linkToEvidence rdf:object ?annotation ;
                        up:attribution ?attribution .
        ?attribution up:source ?source .
        ?source a up:Journal_Citation .
      } 
      GROUP BY ?source ORDER BY DESC(COUNT(?attribution))
